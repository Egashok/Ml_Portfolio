<h4>2.4 Переведем стоимость на современный эквивалент</h4>


```python
from bs4 import BeautifulSoup
import requests

url="http://www.usinflationcalculator.com/inflation/consumer-price-index-and-annual-percent-changes-from-1913-to-2008/"

r=requests.get(url)
data=r.text
soup=BeautifulSoup(data, 'html.parser')

table=soup.find('table')
rows=table.tbody.findAll('tr');

years=[]
cpis=[]

for row in rows:
   
    year=row.findAll('td')[0].get_text()

    if year.isdigit() and int(year)<2017:
        years.append(int(year))
        cpis.append(float(row.findAll('td')[13].get_text()))
cpi_table=pd.DataFrame({
    "year":years,
    "avg_annual_cpi":cpis
})

cpi_table.head()


def get_real_value(nominal_amt,old_cpi,new_cpi):
    real_value=(nominal_amt*new_cpi)/old_cpi
    return real_value


CPI_2016=float(cpi_table[cpi_table['year']==2016]['avg_annual_cpi'])

real_domestic_gross=[]
real_budget_values=[]

for index,row in df.iterrows():
    gross=row['gross']
    budget=row['budget']
    year=row['title_year']
    cpi=cpi_table[cpi_table['year']==int(year)]['avg_annual_cpi'].astype(float)

    real_gross=get_real_value(gross,cpi,CPI_2016)
    real_budget=get_real_value(budget,cpi,CPI_2016)
    real_domestic_gross.append(real_gross)
    real_budget_values.append(real_budget)
    
df['real_domestic_gross']=real_domestic_gross
df['real_budget']=real_budget_values

    
profits=[]
roi_vals=[]

for index,row in df.iterrows():
    profit=row['real_domestic_gross']-row['real_budget']
    budget=row['real_budget']
    num=profit-budget
    den=budget
    
    roi=(num/den)*100
    
    profits.append(profit)
    roi_vals.append(roi)
    
df['profit']=profits
df['roi']=roi_vals
    

cpi_table.to_csv(r'C:\Users\djego\OneDrive\Рабочий стол\OmSTU_ml\data\Shokhin_FIT-221-CPI.csv', index = False)  
```

# **Проверка гипотез**

# *Гипотеза 1: В 21 века бюджеты фильмов выше чем в 20*

Сгруппируем фильмы по году выпуска и найдем средний бюджет.Данную операцию будет выполнять только с фильма где бюджет 
больше 0 т.к. ранее мы значения NaN заменяли на 0.

```python
df_year_budget=df[(df.budget>0) & df.title_year>0].groupby('title_year')['budget'].mean()

print(df_year_budget)

```

Сгруппируем по веку:

```python
s=0

for i in df_year_budget.keys():
    if float(i)<=2000 and not  float(i)== 0:
        s+= df_year_budget[i]
s20=s/75
print(f"В 20 веке:{s20}") #средний рейтинг 20 века

s=0
for i in df_year_budget.keys():
    if float(i)>2000:
        s+= df_year_budget[i]

s21=s/16
print(f"В 21 веке:{s21}")  #средний рейтинг 20 века

print(f'Больше в {round(s21/s20,3)} раз')
```

**Выводы**
Как мы можем увидеть из таблицы, что средний бюджет 21 века больше среднего бюджета 20 века.Бюджеты мы за ранее перевели на современный курс.Гипотеза верна.

# *Гипотеза 2: В 20 веке средний рейтинг фильмов выше чем в 21*

Найдем средний рейтинг фильмов по году выпуска:

```python
df_year_score=df[(df.imdb_score>0) & df.title_year>0].groupby('title_year')['imdb_score'].mean()
print(df_year_score)
```

Найдем средние значения в 20 и 21 веке:

```python
s=0
for i in df_year_score.keys():
    if float(i)<=2000:
        s+= df_year_score[i]
s20=s/75
print(f"В 20 веке:{s20}") #средний рейтинг 20 века

s=0
for i in df_year_score.keys():
    if float(i)>2000:
        s+= df_year_score[i]
s21=s/16
print(f"В 21 веке:{s21}")  #средний рейтинг 20 века
print(f'Разность: {round(s20-s21,4)}')
```

**Выводы:** 
Гипотеза верна.Средний рейтинг фильмов 20 века больше чем средний рейтинг фильмов 21 века.
